```python
import urllib
import time
import requests
from bs4 import BeautifulSoup

#html = "http://www.nature.com/search?date_range=1980-2017&journal=nature&order=relevance&page=3"
#set up the header. It is important to scrap pages to avoid server denying!
headers = {'User-Agent':'Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.11 (KHTML, like Gecko) Chrome/23.0.1271.64 Safari/537.11',
'Accept':'text/html;q=0.9,*/*;q=0.8',
'Accept-Charset':'ISO-8859-1,utf-8;q=0.7,*;q=0.3',
'Accept-Encoding':'gzip',
'Connection':'close',
'Referer':'http://www.baidu.com/link?url=_andhfsjjjKRgEWkj7i9cFmYYGsisrnm2A-TN3XZDQXxvGsM9k9ZZSnikW2Yds4s&amp;amp;wd=&amp;amp;eqid=c3435a7d00006bd600000003582bfd1f'
}
#set up the fixed part for urls
url='http://www.nature.com/'
page=('search?date_range=1980-2017&journal=nature&order=relevance&page=')


titles = []
authorinfo = []
keywords = []
citation_date = []
citation = []
author_names = []
correspondence_author = []
article_type = []

for i in range(51, 70):
    i=str(i)
    a=(url+page+i+'/')
   # req = urllib.request(a, headers={ 'User-Agent': 'Mozilla/5.0' })
   # source = urllib.urlopen(req).read()
    r = requests.get(url=a,headers=headers)
    source = r.content
    soup = BeautifulSoup(source, 'html.parser')
    headings = soup.find_all('h2', attrs={'role': 'heading'})

    for heading in headings:
        titles.append(heading.text)

    soup1 = BeautifulSoup(source, 'lxml')

    jk = soup1.find_all('a', attrs={'data-track-dest': 'Nature'})

    for a in jk:
        source1 = urllib.request.urlopen(a.get('href')).read()
        soup1 = BeautifulSoup(source1, 'html.parser')
        authorinfo.append([e.get_text() for td in soup1.find_all('ol', attrs={'class': 'affiliations'})
                                 for e in td('h3')])
        keywords.append(
                  [e.get_text() for td in soup1.find_all('div', attrs={'class': 'article-keywords inline-list cleared'})
                   for e in td('a')])
        citation_date.append([e.get_text() for td in soup1.find_all('dl', attrs={'class': 'citation dates'})
                                    for e in td('dd')])
        citation.append([e.get_text() for td in soup1.find_all('dl', attrs={'class': 'citation'})
                               for e in td('dd')])
        author_names.append(
                  [e.get_text() for td in soup1.find_all('ul', attrs={'class': 'authors citation-authors'})
                   for e in td('a')])
        correspondence_author.append(
                  [e.get_text() for td in soup1.find_all('ul', attrs={'class': 'correspondence'})
                   for e in td('a')])
        article_type.append([e.get_text() for td in soup1.find_all('div', attrs={'class': 'top-links cleared'})
                                   for e in td('p')])

    time.sleep(0.5)
    print("page"+i+"done !")#print this info to check the progress.


import pandas as pd

db = pd.DataFrame({'titles': titles, 'keywords': keywords, 'citation_date': citation_date,
                   'correspondence_author': correspondence_author,
                   'author_names': author_names, 'authorinfo': authorinfo,
                   'citation': citation, 'article_type': article_type})
db.head()
print(db[:50])
db.to_csv('Nature_Journal_Wepscraping_2.csv', index=False)

```
